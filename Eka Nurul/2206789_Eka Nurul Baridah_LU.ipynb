```python
import numpy as np

# Fungsi jalanLU untuk faktorisasi LU
def jalanLU(A, b):
    n = len(A)
    L = np.zeros((n, n)  # Membuat matriks n x n dengan elemen 0

    for i in range(n):
        L[i][i] = 1  # Mengisi diagonal matriks L dengan 1

    for k in range(n-1):
        # Pertukaran baris jika elemen diagonal utama adalah nol
        if A[k][k] == 0:
            for s in range(n):
                v = A[k][s]
                u = A[k+1][s]
                A[k][s] = u
                A[k+1][s] = v

        for j in range(k+1, n):
            m = A[j][k] / A[k][k]  # Menghitung nilai m
            L[j][k] = m  # Mengisi matriks L dengan nilai m
            for i in range(n):
                A[j][i] = A[j][i] - m * A[k][i]  # Mengurangkan baris A[j] dengan m * baris A[k]

    print('Matriks L')
    print(L)

    print('Matriks U')
    print(A)  # Mencetak matriks U (hasil faktorisasi LU)

    # Mencari vektor y
    y = np.zeros((n, 1)  # Membuat matriks n x 1 dengan elemen 0
    y[0][0] = b[0][0] / L[0][0]  # Menghitung elemen pertama dari vektor y
    for j in range(1, n):
        S = 0
        for i in range(j):
            S = S + y[i][0] * L[j][i]  # Menghitung komponen j dari vektor y
        y[j][0] = b[j][0] - S

    # Mencari solusi vektor x
    x = np.zeros((n, 1)  # Membuat matriks n x 1 dengan elemen 0
    x[n-1][0] = y[n-1][0] / A[n-1][n-1]  # Menghitung elemen terakhir dari vektor x
    for j in range(n-2, -1, -1):
        S = 0
        for i in range(j+1, n):
            S = S + A[j][i] * x[i][0]  # Menghitung komponen j dari vektor x
        x[j][0] = (y[j][0] - S) / A[j][j]

    # Mencetak solusi
    print('Solusi:\n')
    for i in range(n):
        print(f'X{i+1} = {x[i]}')

# Meminta input matriks A dari user
A = []
n = int(input("Masukkan ukuran matriks (n x n): "))
print(f"Masukkan elemen-elemen matriks A ({n}x{n}):")
for i in range(n):
    row = []
    for j in range(n):
        elem = float(input(f"A[{i+1}][{j+1}]: "))
        row.append(elem)  # Membentuk satu persatu baris dari matriks A
    A.append(row)  # Dari baris demi baris membentuk matriks A

# Meminta input matriks b dari user
b = []
print("Masukkan elemen-elemen matriks B:")
for i in range(n):
    elem = float(input(f"B[{i+1}]: "))
    b.append([elem])

# Memanggil fungsi jalanLU dengan matriks A dan b yang diinput oleh user
jalanLU(np.array(A), np.array(b))
```
