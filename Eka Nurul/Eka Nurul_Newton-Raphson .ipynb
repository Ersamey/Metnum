```python
import numpy as np
import matplotlib.pyplot as plt

# Menginisialisasi fungsi untuk metode Newton-Raphson
def newton_raphson(f, df, xi, e):
    print("Nilai xi: ", xi)
    # Mengecek apakah nilai fungsi pada xi sudah mendekati nol
    if abs(f(xi)) < e:
        return xi
    else:
        # Iterasi Newton-Raphson untuk mendapatkan nilai baru xi
        xi = xi - f(xi) / df(xi)
        # Rekursi untuk melanjutkan iterasi sampai mendekati akar
        return newton_raphson(f, df, xi, e)

# Mendefinisikan fungsi f(x) dan turunannya f'(x)
y = input("Masukkan fungsi f(x): ")
fx = lambda x: eval(y)  # Definisi fungsi f(x)
z = input("Masukkan fungsi f'(x): ")
f_prime = lambda x: eval(z)  # Definisi turunan f'(x)
n = float(input("Masukkan Nilai Aproksimasi Awal: "))
e = 1e-6  # Nilai toleransi untuk mendekati nol

# Pemanggilan fungsi Newton-Raphson untuk mencari akar
estimate = newton_raphson(fx, f_prime, n, e)
print("Estimasi akar = %.3f" % estimate)

# Plot fungsi
x = np.linspace(estimate - 5, estimate + 5, 100)  # Membuat array x untuk plotting
y = fx(x)
plt.plot(x, y, label='f(x)')  # Plot fungsi f(x)
plt.axhline(0, color='red', linestyle='--')  # Garis horisontal merah pada y=0
plt.axvline(estimate, color='green', linestyle='--', label=f'Akar (x={estimate:.4f})')  # Garis vertikal hijau pada akar
plt.legend()  # Menampilkan legenda
plt.xlabel('x')  # Label sumbu x
plt.ylabel('y')  # Label sumbu y
plt.title('Grafik Fungsi')  # Judul plot
plt.grid(True)  # Menampilkan grid
plt.show()  # MenampilkanÂ plot
```
