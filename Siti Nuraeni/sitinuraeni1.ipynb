```python
#soal no 1 yang ke b

import numpy as np #panggil library
def my_bisection(f, a, b, e):
  if np.sign(f(a)) == np.sign(f(b)): #np itu numpy jika f(a) == f (b)
    raise Exception('Tidak ada akar pada interval a dan b') #tidak ada akar interval
  m = (a + b)/2 #m itu titik tengah
  if np.abs(f(m)) < e: #jika nilai absolut kurang dari error
    return m # m/c kembalikan nilai m sebagai akar yang ditemukan
  elif np.sign(f(a)) == np.sign(f(m)): #jika fungsi a dan m adalah sama
    return my_bisection(f, m, b, e) #maka memanggil my bisection
  elif np.sign(f(b)) == np.sign(f(m)):#jika fungsi b dan m adalah sama
    return my_bisection(f, a, m, e) #maka memanggil my bisection

e = 2.71828182845904 #nilai dari e^-1
f = lambda x: e**x - x #soal fungsi metode bagi dua

r1 = my_bisection(f,-1,0, 0.1) #panggil bisection yang galat nya 0.1
print("r1 =", r1) #print hasil r1
print("f(r1) =", f(r1)) #print f (r1)

r01 = my_bisection(f, -1,0, 0.01) #panggil bisection yang galat nya 0.01
print("r01 =", r01) #print hasil r01
print("f(r01) =", f(r01)) #print f(r01)
```
