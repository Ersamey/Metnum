```python
import numpy as np
import matplotlib.pyplot as plt

#Menginisialisasi fungsi untuk menghitung f(xi), f'(xi), xi, dan e
def newton_raphson(f, df, xi, e):
    print("Nilai xi: ", xi) #mencetak nilai xi
    if abs(f(xi)) < e: #jika f(xi) < e (toleransi)
        return xi #maka menjadi nilai xi
    else: #lainnya jika
        xi = xi - f(xi) / df(xi) #ubah nilai xi menggunakan newton-rapson
        return newton_raphson(f, df, xi, e) #panggil newton-raphson yang telah diperbarui

y = input("Masukkan fungsi f(x): ")
fx = lambda x: eval(y)  #masukkan f(x)
z = input("Masukkan fungsi f'(x): ")
f_prime = lambda x: eval(z)  #masukkan f'(x)
n = float(input("Masukkan Nilai Aproksimasi Awal: "))

#panggil fungsi Newton-Raphson
estimate = newton_raphson(fx, f_prime, n, e)
print("estimate = %.3f" % estimate)

# Plot fungsi
x = np.linspace(estimate - 5, estimate + 5, 100)  #Membuat array x untuk plotting
y = fx(x) #nilai y pad setiap titik x
plt.plot(x, y, label='f(x)')  #Plot fungsi f(x)
plt.axhline(0, color='red', linestyle='--')  #Garis horizontal merah pada y=0
plt.axvline(estimate, color='green', linestyle='--', label=f'Akar (x={estimate:.4f})')  #Garis vertikal hijau pada akar
plt.legend()  #Menampilkan legenda
plt.xlabel('x')  #Label sumbu x
plt.ylabel('y')  #Label sumbu y
plt.title('Grafik Fungsi')  #Judul plot
plt.grid(True)  #Menampilkan grid
plt.show()  #MenampilkanÂ plot


```
