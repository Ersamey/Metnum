```python
#soal nomor 2 yang ke-1
import numpy as np

def my_bisection(f, a, b, e, iterasi=0, iterasimax=3):
    if np.sign(f(a)) == np.sign(f(b)): #jika tanda f(a) = tanda f(b)
        raise Exception('Tidak ada akar pada interval a dan b') #maka tidak ada akar interval

    c = (a + b) / 2 #nilai tengah
    if np.abs(f(c)) < e or iterasi >= iterasimax: #jika absolut c kurang dari eror atau iterasi lebih dari sama dengan iterasi maksimal
        return c #maka memanggil c

    if np.sign(f(a)) == np.sign(f(c)): #jika f(a) sama dengan f(c)
        return my_bisection(f, c, b, e, iterasi+1, iterasimax) #maka memanggil bisection
    elif np.sign(f(b)) == np.sign(f(c)): #lainnya jika f(b) sama dengan f(c)
        return my_bisection(f, a, c, e, iterasi+1, iterasimax) #maka memanggil bisection jga

f = lambda x: x**3 - 2*x + 1 #soal fungsi metode bagi dua

r1 = my_bisection(f, -2, -1, 0.01) #panggil bisection dengan galat 0.01
print("r1 =", r1) #maka print r1 nilai akar
print("f(r1) =", f(r1)) #print f(r1)
```

```python
#soal no.2 yang ke-3

import numpy as np
import matplotlib.pyplot as plt

def my_bisection(f, a, b, e, iterasi=0, iterasimax=3):
    if np.sign(f(a)) == np.sign(f(b)): #jika tanda f(a) = tanda f(b)
        raise Exception('Tidak ada akar pada interval a dan b') #maka tidak ada interval

    c = (a + b) / 2 #nilai tengah
    if np.abs(f(c)) < e or iterasi >= iterasimax: ##jika absolut c kurang dari eror atau iterasi lebih dari sama dengan iterasi maksimal
        return c #memanggil nilai c

    if np.sign(f(a)) == np.sign(f(c)): #jika f(a) sama dengan f(c)
        return my_bisection(f, c, b, e, iterasi+1, iterasimax) #maka memanggil bisection
    elif np.sign(f(b)) == np.sign(f(c)): #jika f(a) sama dengan f(c)
        return my_bisection(f, a, c, e, iterasi+1, iterasimax) #maka memanggil bisection

# memasukan fungsi dari user
expression = input("Masukkan fungsi f(x): ")
f = lambda x: eval(expression)

# masukan a,b, dan galat nya
a = float(input("Masukkan batas awal (a): "))
b = float(input("Masukkan batas akhir (b): "))
e = float(input("Masukkan galat (e): "))
iterasimax = int(input("Masukkan maksimum iterasi: ")) #maksimal 3

# Hitung akar menggunakan metode bisection
r1 = my_bisection(f, a, b, e, iterasimax=iterasimax)
print("Akar =", r1)
print("f(r1) =", f(r1))

# Plot fungsi
x = np.linspace(a, b, 100) #titik x grafik
y = f(x) #hitung nilai y pada titik x
plt.plot(x, y, label='f(x)') #plot garis menggunakan titk x,y yang telah dihitung sebelumnya
plt.axhline(0, color='red', linestyle='--') #menambahkan garis horizontal pada y dan warna red. linestyle memberikan garis -- untuk sumbu x
plt.axvline(r1, color='green', linestyle='--', label=f'Akar (x={r1:.4f})') #garis vertikal yang menunjukkan akar dan telah ditemukan pada r1
plt.legend() #menampilkan legenda grafik
plt.xlabel('x') #label sumbu x
plt.ylabel('y') #label sumbu y
plt.title('Grafik Fungsi') #judul grafik fungsi
plt.grid(True) #menampilkan grid grafik
plt.show() #menampilkan grafik secara keseluruhan
```
